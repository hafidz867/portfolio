```python
def min_edit_distance():
  word_A = input("Please input your first word: ")
  word_B = input("Please input your second word: ")
  word_A = word_A.lower()
  word_B = word_B.lower()
  len_A = len(word_A)
  len_B = len(word_B)
  D = [[0 for x in range(len_B+1)] for x in range(len_A+1)]
  for i in range(len_A+1):
    for j in range(len_B+1):
      print("\nIterasi i={}, j={}".format(i,j))
      if i == 0:
        D[i][j] = j
        print("D[{}][{}] = {}".format(i,j,D[i][j]))
      elif j == 0:
        D[i][j] = i
        print("D[{}][{}] = {}".format(i,j,D[i][j]))
      elif word_A[i-1] == word_B[j-1]:
        D[i][j] = D[i-1][j-1]
        print("D[{}][{}] = {}".format(i,j,D[i][j]))
      else:
        D[i][j] = 1 + min(D[i][j-1],D[i-1][j],D[i-1][j-1])
        print("D[{}][{}] = {}".format(i,j,D[i][j]))
        
  return "\nMinimum Edit Distance: {}".format(D[len_A][len_B])
```

```python
print(min_edit_distance())
```
