##**Training Data**

```python
#import data training

df_training = pd.read_csv('twitter_training.csv', names=['ID', 'Entity', 'Sentiment', 'Text'])
df_training = df_training[['Text', 'Sentiment']]
df_training = df_training.dropna()
df_training
```

```python
#we only use 50% of data training which will be trained 

df_training = df_training.sample(frac=0.5)
df_training.reset_index(drop=True, inplace=True)
```

```python
df_training['Sentiment'].value_counts()
```

```python
#adding new dimensions with textblob (polarity and subjectivity)

def tb_enrich(ls):
    tb_polarity = []
    tb_subjectivity = []
    for tweet in ls:
        tb_polarity.append(TextBlob(tweet).sentiment[0])
        tb_subjectivity.append(TextBlob(tweet).sentiment[1])
    return tb_polarity, tb_subjectivity

df_training["Polarity"], df_training["Subjectivity"] = tb_enrich(list(df_training["Text"]))
df_training
```

```python
#change Sentiment column into number that represent its categories

category_to_index = {"Neutral":0, "Irrelevant":0, "Negative":2, "Positive": 3}
names_to_ids = lambda n: np.array([category_to_index.get(x) for x in n])
df_training["Sentiment"] = names_to_ids(df_training["Sentiment"])
df_training
```

```python
def remove_stopwords(lemma):
    #lemmatises and then removes stopwords
    lemma = [lemmatiser.lemmatize(word) for word in lemma if word not in (stop_english) and (word.isalpha())]
    lemma = " ".join(lemma) #joins the words back into a single string
    return lemma

#slits each string into a list of words using tokenizer
df_training["Text_Split"] = df_training["Text"].apply(word_tokenize)

lemmatiser = WordNetLemmatizer()
stop_english = Counter(stopwords.words())
df_training["Text_Split"] = df_training["Text_Split"].apply(remove_stopwords)
```

```python
df_training.head()
```

```python
#define the Tokeniser
tokeniser = Tokenizer(num_words=1000, lower=True)

#ceate the corpus by finding the most common 
tokeniser.fit_on_texts(df_training["Text_Split"])
```

```python
text_tokens = tokeniser.texts_to_matrix(list(df_training["Text_Split"]))
```

```python
#combining the dataframe with the tokens
full_df = pd.concat([df_training, pd.DataFrame(text_tokens)], sort=False, axis=1)
full_df.shape
```

```python
#remove the sentiment into new variable
sentiment = full_df["Sentiment"]

#drop all non-useful columns
full_df = full_df.drop(["Sentiment", "Text", "Text_Split"], axis=1)
```

```python
full_df
```

```python
p.set_options(p.OPT.MENTION, p.OPT.EMOJI, p.OPT.HASHTAG, p.OPT.RESERVED, p.OPT.SMILEY, p.OPT.URL)
nltk.download('words')
words = set(nltk.corpus.words.words())
```

```python
def cleaner(text):
    text = re.sub("@[A-Za-z0-9]+", "", text) #Remove @ sign
    text = ''.join(c for c in text if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    text = text.replace("#", "").replace("_", "") #Remove hashtag sign but keep the text
    text = " ".join(w for w in nltk.wordpunct_tokenize(text) \
         if w.lower() in words or not w.isalpha()) #Remove non-english tweets (not 100% success)
    return text
```

```python
def translater(text):
  text = GoogleTranslator(source='auto',target='en').translate(text)
  return text
```

```python
def sentiment(text):
  blob = TextBlob(str(text))
  score = blob.polarity
  return score
```

```python
def get_sentiment(sentiment_score):
   if sentiment_score == 0:
    return 'Neutral/Irrelevant'
   elif sentiment_score == 2:
    return 'Negative'
   else:
    return 'Positive'  
```
