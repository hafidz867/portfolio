# Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Input Dataset

```python
df = pd.read_csv('e-shop clothing 2008.csv', delimiter=';')
df
```

# Analisis Dataset

```python
fig = plt.figure(figsize=(16, 4))
ax = fig.add_subplot(111)

data_boxplot = [df['country'], df['colour'], df['price']]

boxplot = ax.boxplot(data_boxplot, patch_artist=True, notch='True', vert=0)
 
colours = ['#ebeb1e', '#1eeb4e', '#1e32eb']
 
for patch, colour in zip(boxplot['boxes'], colours):
    patch.set_facecolor(colour)
    
# mengubah warna pada garis median
for median in boxplot['medians']:
    median.set(color='black', linewidth=2)
     
# Label pada x-axis
ax.set_yticklabels(['country', 'colour', 'price'])
 
# judul
plt.title("Boxplot Analysis")
plt.grid()
plt.show(boxplot)
```

```python
fig = plt.figure(figsize=(16, 2))
ax = fig.add_subplot(111)

boxplot = ax.boxplot(df['order'], patch_artist=True, notch='True', vert=0)
 
colours = ['#e72929']
 
for patch, colour in zip(boxplot['boxes'], colours):
    patch.set_facecolor(colour)
    
# mengubah warna pada garis median
for median in boxplot['medians']:
    median.set(color='black', linewidth=2)

# label pada x-axis
ax.set_yticklabels(['order'])
 
# judul
plt.title("Boxplot Analysis")
plt.show(boxplot)
```

```python
fig = plt.figure(figsize=(7, 15))
plt.scatter(df['month'], df['country'])

xticks_num = [4, 5, 6, 7, 8]
xticks_str = ['April', 'May', 'June', 'July', 'August']
yticks = np.arange(1,48).tolist()

plt.grid()
plt.xticks(xticks_num, xticks_str)
plt.yticks(yticks)
plt.title('Month to Country Plot')
plt.xlabel('\nMonth')
plt.ylabel('Country Code')
plt.show()
```

```python
order = df['order']
order_x = ['<= 40', '41 - 80', '81 - 120', '121 - 160', '> 160']
order_y = [sum(order<=40), 
           sum((order>40) & (order<=80)), 
           sum((order>80) & (order<=120)), 
           sum((order>120) & (order<=160)), 
           sum(order>160)]
```

```python
def labels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]+2000, y[i], ha = 'center')
plt.figure(figsize=(10,10))
plt.bar(order_x, order_y)
labels(order_x, order_y)
plt.title('Order Count Plot')
plt.xlabel('Order Quantity')
plt.ylabel('Amount of Transactions')
plt.show()
```

```python
def labels(x,y):
    for i in range(len(x)):
        plt.text(i+4, y[i]+5000, y[i], ha = 'center')

fig = plt.figure(figsize=(7, 5))

month = df['month'].value_counts().sort_index().keys().tolist()
order_per_month = df.groupby('month')['order'].sum()

xticks_str = ['April', 'May', 'June', 'July', 'August']

plt.bar(month, order_per_month)
labels(month, order_per_month.tolist())
plt.xticks(month, xticks_str)
plt.title('Order per Month Plot')
plt.xlabel('\nMonth')
plt.ylabel('Order')
plt.show()
```

```python
fig = plt.figure(figsize=(16,10))
plt.scatter(np.arange(df['order'].count()), df['order'])
plt.grid(axis='y')
plt.show()
```

```python
fig = plt.figure(figsize=(16,10))
april = df['order'][df['month']==4].count()
plt.scatter(np.arange(april), df['order'][df['month']==4])
plt.title('Total Order April')
plt.show()
```

```python
def labels(x,y):
    for i in range(len(x)):
        plt.text(i+1, y[i]+15000, y[i], ha = 'center')

fig = plt.figure(figsize=(16, 10))

country = df['country'].value_counts().sort_index().keys().tolist()
order_per_country = df.groupby('country')['order'].sum()

plt.scatter(country, order_per_country)
plt.plot(country, order_per_country)
labels(country, order_per_country.tolist())
plt.xticks(country)
plt.title('Order per Country')
plt.xlabel('\nCountry Code')
plt.ylabel('Amount of Order')
plt.show()
```

```python
order_day_one = df['order'][df['month']==4][df['day']==1]

fig = plt.figure(figsize=(16,10))
plt.scatter(np.arange(order_day_one.count()), order_day_one)
plt.show()
```

```python
order_day_x = []
for i in df['month'].value_counts().sort_index().keys().tolist():
    for j in df['day'].value_counts().sort_index().keys().tolist():
        order_day_x.append(df['order'][df['month']==i][df['day']==j].sum())
order_day_x
```

```python
fig = plt.figure(figsize=(16,10))

plt.scatter(np.arange(len(order_day_x)), order_day_x)
plt.xticks(np.arange(0,len(order_day_x),5))
plt.yticks(np.arange(0,34000,1000))
plt.grid()
plt.title('Order per Day')
plt.xlabel('\nDay')
plt.ylabel('Amount of Order')
plt.show()
```
